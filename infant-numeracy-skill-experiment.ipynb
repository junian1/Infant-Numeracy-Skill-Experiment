## Hypotheses
### One-to-one Principle
Infants in the multilingual will show a siginificantly stronger preference for one-to-one principle consistent counting compared to infants in the monolingual condition
<br>
### Cardinality Principle (English) 
Infants in the multilingual will show a siginificantly stronger preference for English cardinality principle consistent counting compared to infants in the monolingual condition 
<br>
### Cardinality Principle (Extinct)
Infants in the multilingual will show a siginificantly stronger preference for extinct language cardinality principle consistent counting compared to infants in the monolingual condition
<br>
### Stable Order Principle
Infants in the multilingual will show a siginificantly stronger preference for stable order principle consistent counting compared to infants in the monolingual condition

## Method
### Design
This study is a 1 IV (2 levels) and 4 DVs design

### Variables
#### Independent Variable (IV) 
1. Type of counting book (Condition)<br>
<br>
<i>1 = Monolingual<br>
   2 = Multilingual </i>
<br>
<br>

#### Counting Books
In the counting books, there were 10 pages of various types of animals. On each page, there were 5 to 7 animals with counting words below the animals. Parents could read to their infants based on the counting words shown below each animal. In the English counting book, only English counting words were used, "One, Two, Three, Four, Five, Six, Seven". In the multilingual counting book, the counting words were replaced by counting words in other languages, and each page had a different language (Spanish, French, Mandarin…).
<br>
<br>
#### Dependent Variables (DVs)
1. Preference for One-to-one Principle Counting in Extinct Language (CorIncor)
1. Preference for Cardinality Principle Counting in English (EngTotal)
1. Preference for Cardinality Principle Counting in Extinct Language (ExtTotal)
1. Preference for Stable Order Principle Counting in Extinct Language (YellowGreen)<br>
<br>
<i>0 = Average<br>
    Positive Number = Preference for Principle Consistent Counting<br>
    Negative Number = Prefernece for Principle Inconsistent Counting</i>
<br>
<br>

#### One-to-one Principle
Infants watched a total of 8 videos, 4 in one-to-one principle consistent counting (Correct1, Correct2, Correct3, Correct4) and 4 in one-to-one principle inconsistent counting (Incorrect1, Incorrect2, Incorrect3, Incorrect4). Each video shows a 22 seconds animation of a female counting fish in an extinct language (eg. "One, Two, Three, Four, Five, Six") repeated twice (11 seconds each round).
In the one-to-one principle consistent counting videos, the female points to each fish as she counts. Whereas in the one-to-one principle inconsistent counting videos, the female points at 2 fish only as she counts. Infants were recorded during the session, and the total viewing duration was calculated based on how long infants watched the counting videos.
<br>
<br>
#### Cardinality Principle (English & Extinct)
6 toys were placed inside a black box, and a researcher secretly took out 2 toys from a hidden back opening. Infants were asked to search what was inside the box from a front opening, and a 10 seconds search time began (EngTime1, ExtTime1). After 4 toys have been taken out, the researcher discreetly put the 2 toys back from the back opening. The researcher then reached to the front opening and takes out the other 2 toys. Another 10 seconds search time begins to measure whether the infant would still search inside the box (EngTime2, ExtTime2). This process was repeated twice, once in English and once in an extinct language.
<br>
<br>
#### Stable Order Principle
2 buttons were placed in front of the infants, a yellow button (YellowButton) and a green button (GreenButton). The yellow button will trigger an animated video of a female counting fish in stable order principle consistent counting using an extinct language (eg."One, Two, Three, Four, Five, Six"). The green button will trigger an animated video of a female counting fish in stable order principle inconsistent counting using an extinct language in a random order (eg. "Four, Two, One, Five, Six, Three"). The YellowButton and GreenButton show the number of times infants pressed each button.
<br>
<br>

## Summary of Results
This research has 4 main (a priori) hypotheses:
1. Infants in the multilingual condition will show a siginificantly stronger preference for one-to-one principle consistent counting compared to infants in the monolingual condition
1. Infants in the multilingual condition will show a siginificantly stronger preference for English cardinality principle consistent counting compared to infants in the monolingual condition
1. Infants in the multilingual condition will show a siginificantly stronger preference for extinct language cardinality principle consistent counting compared to infants in the monolingual condition
1. Infants in the multilingual condition will show a siginificantly stronger preference for stable order principle consistent counting compared to infants in the monolingual condition
<br>
<br>

The T-Test reveals that infants who were in the mulitlingual condition did not show significantly stronger preference for all but 1 principle consistent counting, which is stable order principle. However, further analysis using one-sample t-test suggest that their preference is not siginficantly above chance. Our research suggests that exposure to multiple languages does not accelerate infant's understanding of all 3 counting principles compared to exposure to counting in 1 language only.
<br>
<br>

3 potential reasons this study did not find any significant effect
1. Infants have yet to form abstract understanding of the counting principle at this age, their understanding of counting is only limited to counting in their own language but not the principles, which explains why they failed to transfer their understanding to other languages. 
<br>
<br>
In order to test this hypothesis, we could add another series of tests on one of the 10 languages used in the counting book. If the hypothesis is true, only infants in the mulitlingual condition would show a significantly stronger preference for the counting in that language.
<br>
<br>
1. The book reading intervention failed to teach infants the counting principle on an abstract level, and a stronger intervention is needed to cause such an effect. Since the intervention used in this research was 10 languages for 4 weeks on 14-18 months old infants. Perhaps varying the number of languages, duration and infant's age could generate a significantly stronger effect.
<br>
<br>
A series of A/B tests on varying the different counting book is needed to find the ideal intervention that can produce the strongest effect of accelerating infants' understanding of the counting principles.
<br>
<br>
1. This research is hugely underpowered compared to similar past research. Among similar past research with low sample size, only cardinality principle was detected to show significant results. This could be true since stable order principle nearly found significant with p = 0.053. Increasing the sample size could help increase power - the probability to correctly reject the null hypothesis

## Data Wrangling & Feature Engineering

```python
#Add new column (Days) to calculate infant's age in days

df_participant['DateofTest'] = pd.to_datetime(df_participant['DateofTest'])
df_participant['DOB'] = pd.to_datetime(df_participant['DOB'])

df_participant['Days']=df_participant['DateofTest']- df_participant['DOB']
df_participant.head()
```

```python
#Add new column (Months) to calculate infant's age in months
df_participant['Months']=round(df_participant.Days/np.timedelta64(1,'M'),2)
df_participant.head()
```

```python
#Convert 'Days' to numeric type
df_participant['Days']=pd.to_numeric(df_participant['Days'])
```

```python
#Standardize scores (percentage) for one-to-one principle data
df_experiment['CorrectPerc']=df_experiment['CorrectTotal']/88
df_experiment['IncorrectPerc']=df_experiment['IncorrectTotal']/88
df_experiment['CorIncorPercDif']=df_experiment['CorrectPerc']-df_experiment['IncorrectPerc']
df_experiment['CorIncorPerc']=df_experiment['CorIncorPercDif']/(df_experiment['CorrectPerc']+df_experiment['IncorrectPerc'])
df_experiment.head()
```

```python
#Add a new column (EngExt) to measure difference in Cardinality pricinple scores between English and Extinct Language
df_experiment['EngExt']=df_experiment['EngTotal']-df_experiment['ExtTotal']
df_experiment.head()
```

```python
#Add a new colum (YellowGreenPerc) to measure preference for yellow vs green button in stable order principle
df_experiment['YellowGreenPerc']=df_experiment['YellowGreen']/(df_experiment['YellowButton']+df_experiment['GreenButton'])
df_experiment.head()
```
